### 7. For each book in `goodbooks`, compute the list of its authors. Assuming that the number of reviews with a text (column `work_text_reviews_count`) is split equally among all authors, find for each authors the total number of reviews with a text. We will call this quantity the *shared number of reviews with a text*.

Gli autori del medesimo libro sono separati da una virgola, utilizzeremo dunque la virgola come conteggio del numero di autori presenti per ogni libro.

```python
gb_books['number_authors'] = gb_books['authors'].str.count(',')
gb_books.head(15)
```

```python
gb_books['number_authors'] = gb_books['number_authors'] + 1
gb_books.head(15)
```

Eseguiamo una split degli autori sulle virgole:

```python
gb_books['author_name'] = gb_books['authors'].str.split(',')
gb_books.head(15)
```

Utilizziamo il metodo explode per dividere le righe che presentano più di un autore.

```python
gb_books_authors = gb_books.explode('author_name')
gb_books_authors.head()
```

```python
gb_books_authors['splitted_reviews'] = gb_books_authors['work_text_reviews_count'] / gb_books['number_authors']
gb_books_authors.head(15)
```

```python
gb_books_authors.groupby('author_name', as_index = False)['splitted_reviews'].sum()
```
