L'obiettivo della visualizzazione consiste nel mostrare se il numero di score postoperatori fisici che migliorano o peggiorano è influenzato dalla temperatura media.

```python
#importazione delle librerie
import pandas as pd
import plotly.io as pio
import plotly.graph_objs as go
```

Importazione dei dati

```python
dataset= pd.read_csv("./compitino_dataviz_2021_2022.csv")
dataset.head()
```

Creazione di nuove colonne nel dataset utili alla visualizzazione

```python
dataset["delta_score"] = dataset["score_postop"] - dataset["score_preop"] 
dataset["peggiorato"] = dataset["delta_score"] < 0
dataset["pshysical_score"] = dataset["nome_modulistica"] == "SF12 PhysicalScore"
```

Visualizzazione dei dati mediante utilizzo di violin plot e annesso scatter plot dei punteggi che sono migliorati o peggiorati in relazione alla temperatura media.
Si può notare come il numero di punteggi migliorati è distribuito più uniformemente rispetto al numero di punteggi peggiorati

```python
import plotly.express as px
import plotly.graph_objects as go
fig = go.Figure();
fig.add_trace(go.Violin(x= dataset["peggiorato"],y= dataset["temperatura_media"]))
fig.update_traces(points="all")
fig.update_layout(go.Layout(plot_bgcolor='rgba(0,0,0,0)', violinmode='group',
                  xaxis_title="Score peggiorato", yaxis_title="Temperatura media",
                  xaxis= dict(tickmode= "array", tickvals =[True, False], ticktext = ["Peggiorato", "Migliorato"]), yaxis={"mirror" : True, 'side': 'right'}))
fig.show()
```

Salvataggio della visualizzazione

```python
import plotly.io as pio
pio.kaleido.scope.default_format = "png"
pio.kaleido.scope.default_width = 1700

# fig.to_image(format="png", engine="kaleido", width=1700)
fig.write_image("violins.png", scale=1, width= 1700)
```
